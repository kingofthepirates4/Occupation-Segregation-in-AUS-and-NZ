## Understanding the raw data and the differences between both datafiles

### Australia's datafile


**Format:**

Australia's datafiles have 302 columns in total only one row (excluding the header).

The columns are in the general form of *P/M/F_{occupation_code}_{work_hour_range}* or *P/M/F/_{occuption_code}_total* where P stands for persons, M for male, F for female and T for total.

The only exceptions of this format is  *P/M/F/T_Tot_{work_hour_range}* which represents the the total male/female/persons working a work hour range across all occupations.

**Occupation Categories:**

Australia's data uses short occupation codes to refer to these occupation categories. The mapping is as follows: 
                                                                                                                                      
* Mng: Manager

* Pro: Professionals

* TTW: Technicians and trades workers

* CPS: Community and personal service workers

* CA: Clerical and administrative workers

* Sal: Sales workers

* MOD: Machinery operators and drivers

* Lab: Labourers

Apart from the occupation categories mentioned above, the datafile contains also contains a category 'ID/NS' (Inadequately described/Not stated) which refers to individuals whose occupations are either undefined or not reported in the census data.

**Work Hour Ranges:**

Australia's work hours categories are in different ranges from 0 to 50+ with one exception, that is, the 'NS' (Not stated).

Australia's work hour ranges: 0, 1-19, 20-29, 30-34, 35-39, 40-44, 45-49, 50+, NS (Not stated).
                                                                                                                                        
### New Zealand's datafile

**Format:**

There are totally 330 rows and 18 columns, out of which 4 columns contain NaN or no values ('Year', 'Observation value', 'OBS_STATUS', 'Observation status'), and another 4  are columns that are of major concern for this analysis ('Hours worked in employment per week', 'Occupation', 'Sex' and 'OBS_VALUES').

**Occupation Categories:**

New Zealand's occupation categories are the same as its Australian counterpart ('Managers', 'Professionals', 'Technicians and Trades Workers', 'Community and Personal Service Workers', 'Clerical and Administrative Workers', 'Sales Workers', 'Machinery Operators and Drivers', 'Labourers'). 

New Zealand's data does not use short occupation codes to refer to these occupations.

**Work Hour Ranges:**

New Zealand's work hour ranges: 1-9, 10-19, 20-29, 30-39, 40-49, 50-59, 60+

## Data Cleaning and Processing

Both the datafiles are in different formats, have varying number of rows and columns, have different occupation categories and different work ranges. Therefore, it is necessary to clean the files, process them and organize them in a uniform data structure, in order to compare and analyse them. 
1. The files need to be checked for missing or invalid data
2. The work hours and occupation categories need to be dynamically read and extracted.
3. Uniform categories/headers need to be formed
    * The profession codes in Australia's datafile should be mapped to their full forms.
    * The work hour ranges need to be sorted.
    * Extraneous strings such as 'hours worked' in New Zealand's file need to be removed.
3. Extracted and cleaned data from both files need to be organized in the following data structures for easy analysis and comparison:
    * Profession dictionary: This dictionary maps each occupation to a DataFrame containing the following four columns:
        * Hours_worked: The range of hours worked.
        * People: The total number of people working those hours.
        * Male: The number of males working those hours.
        * Female: The number of females working those hours.
    * Total dictionary: This dictionary stores the total hours worked by males and females, aggregated across all occupations.
    * Averages dictionary: This will contain the average hours worked of Male, Female and People working in different occupations.
    * Gender Percentage dictionary: This dictionary will store the percentage representation of males and females in each occupation.

```python

# Define new hours worked categories mapping for New Zealand
new_hours_categories_nz = {
    '1-19': ['1-9', '10-19'],
    '20-29': ['20-29'],
    '30-39': ['30-39'],
    '40-49': ['40-49'],
    '50+': ['50-59', '60 hours or more worked']
}

# Creates the profession dictionary for Australia
def organize_aus_professions_dict(df1, df2):
    profession_dfs = {}
    # Extracts set of occupation by splitting column names and excluding 'Tot' and 'ID' categories
    professions = {col.split('_')[1] for col in df1.columns if col.startswith('P_') and col.split('_')[1] not in ['Tot', 'ID']}
    for profession in professions:
        sorted_work_hours = extract_sorted_work_hours(df1, profession)
        profession_standard_name = standard_names(profession)
        profession_dfs[profession_standard_name] = gather_data_for_profession(profession, df1, df2, sorted_work_hours)
    return profession_dfs

# Creates total dictionary for Australia
def create_aus_tot_dataframe(df1, df2):
    # Sorts work hours groups for 'Tot' profession
    sorted_work_hours = extract_sorted_work_hours(df1, 'Tot')  
    # Calls helper function
    return gather_data_for_profession('Tot', df1, df2, sorted_work_hours)


# HELPER FUNCTION
# This function extracts and sorts work hours dynamically 
def extract_sorted_work_hours(df, profession):
    work_hours = set()
    for col in df.columns:
        if col.startswith(f"P_{profession}_"):
            hours_worked = col.split(f"P_{profession}_")[1]
            work_hours.add(hours_worked)
    def hours_worked_sort_key(hours_worked):
        if hours_worked == 'Tot':
            return float('inf') 
        elif hours_worked == 'hours_NS':
            return float('inf') - 1  
        else:
            try:
                parts = hours_worked.split('_')
                return float(parts[0])  
            except ValueError:
                return float('inf')  
    sorted_work_hours = sorted(work_hours, key=hours_worked_sort_key)
    return sorted_work_hours
    
# HELPER FUNCTION
# Forms data rows
def gather_data_for_profession(profession, df1, df2, sorted_work_hours):
    data_rows = []
    for hours_worked in sorted_work_hours:
        male_col = f"M_{profession}_{hours_worked}"
        female_col = f"F_{profession}_{hours_worked}"
        person_col = f"P_{profession}_{hours_worked}"
        if male_col in df2.columns and female_col in df2.columns and person_col in df1.columns:
            male_value = df2[male_col].values[0]
            female_value = df2[female_col].values[0]
            person_value = df1[person_col].values[0]
            data_rows.append({
                'Hours_worked': hours_worked,
                'Male': float(male_value),
                'Female': float(female_value),
                'Person': float(person_value)
            })
    
    return pd.DataFrame(data_rows)
    
# HELPER FUNCTION
# Since there is data discrepency in occupation labels, this function maps abbreviate labels to their full form
def standard_names(profession):
    profession_short_form = ["Mng","Pro","TTW","CPS", "CA", "Sal", "MOD","Lab"]
    profession_full_form = [ "Managers", "Professionals", "Technicians and Trades Workers", "Community and Personal Service Workers","Clerical and Administrative Workers","Sales Workers", "Machinery Operators and Drivers", "Labourers"]
    index = profession_short_form.index(profession)
    return profession_full_form[index]

occupation_short_forms = {
        'Managers': 'Mng',
        'Professionals': 'Pro',
        'Technicians and Trades Workers': 'TTW',
        'Community and Personal Service Workers': 'CPS',
        'Clerical and Administrative Workers': 'CA',
        'Sales Workers': 'Sal',
        'Machinery Operators and Drivers': 'MOD',
        'Labourers': 'Lab'
    }

auss = organize_aus_professions_dict(aus1,aus2)
auss
```

### Organizing New Zealand's data into a professions dictionary and total dictionary

```python
# Creates the profession dictionary for New Zealand.
def organize_nz_professions_dict(df):
    profession_dfs = {}
    professions = df['Occupation'].unique()
    df['Hours worked in employment per week'] = df['Hours worked in employment per week'].apply(clean_hours_worked_label)
    hours_worked_categories = df['Hours worked in employment per week'].unique()
    for profession in professions:
        data_rows = []
        if (profession == 'Total people stated') or (profession == 'Total people - occupation'):
                continue
        profession_df = df[df['Occupation'] == profession]
        for hours_worked in hours_worked_categories:
            if hours_worked == 'Total people stated':
                continue
            hours_df = profession_df[profession_df['Hours worked in employment per week'] == hours_worked]
            male_value = hours_df[hours_df['Sex'] == 'Male']['OBS_VALUE'].sum()
            female_value = hours_df[hours_df['Sex'] == 'Female']['OBS_VALUE'].sum()
            person_value = hours_df[hours_df['Sex'] == 'Total people - sex']['OBS_VALUE'].sum()
            if male_value > 0 or female_value > 0 or person_value > 0:
                if hours_worked == 'Total people':
                    hours_worked = 'Tot'
                data_rows.append({
                    'Hours_worked': hours_worked,
                    'Male': male_value,
                    'Female': female_value,
                    'Person': person_value
                })
        if data_rows:  
            organized_df = pd.DataFrame(data_rows)
            profession_dfs[profession] = organized_df
    return profession_dfs

# Creates total dictionary for New Zealand.
def create_nz_tot_dataframe(nz_data):
    nz_total_people_data = nz_data[nz_data['Occupation'] == 'Total people stated']
    data = []
    for _, row in nz_total_people_data.iterrows():
        # Calls helper function to remove the 'Hours worked in employment per week' label
        work_hours = clean_hours_worked_label(row['Hours worked in employment per week'])
        # Checks if the row is already in the list, if not then adds it
        existing_entry = next((item for item in data if item['Hours_worked'] == work_hours), None)
        if not existing_entry:
            data.append({'Hours_worked': work_hours, 'Male': 0, 'Female': 0, 'Person': 0})
        for entry in data:
            if entry['Hours_worked'] == work_hours:
                if row['Sex'] == 'Male':
                    entry['Male'] = row['OBS_VALUE']
                elif row['Sex'] == 'Female':
                    entry['Female'] = row['OBS_VALUE']
                elif row['Sex'] == 'Total people - sex':
                    entry['Person'] = row['OBS_VALUE']
    df = pd.DataFrame(data)
    return df

# HELPER CLEANING FUNCTION
# Cleans the 'Hours worked' column by removing the trailing 'hours worked' string and other unnecessary labels.
# Removes any non-numeric characters from the start and end of the string
def clean_hours_worked_label(hours_worked):
    if "hours worked" in hours_worked:
        return hours_worked.replace(" hours worked", "").strip()
    elif "hours in employment per week" in hours_worked:
        return "Total people"
    return hours_worked
```

### Creating an average hours worked dictionary
The **average hours worked per profession** is calculated using a weighted average approach, dynamically handling work hour ranges provided in the dataset. In addition to calculating the median hours worked for each group, special cases such as people who worked **0 hours** or whose hours are **not stated** are subtracted from the total population to ensure accurate results.

Calculation Steps:
1. **Dynamic work hours Ranges**:
   - The dataset provides flexible work hour ranges (e.g., `1-9`, `10-19`, `20-29`). For each range, the **median hours worked** is calculated as the midpoint. For example, the midpoint of the range `1-9` is `5`.


2. **Weighted Sum of Work hours**:
   - For each age group, the **median work hours** is multiplied by the number of people in that group. This creates a weighted contribution of each group to the overall work hours.


3. **Exclusion of 0 Hours Worked and Not Stated Hours**:
   - Individuals who reported **0 hours worked** or whose hours are **not stated** (e.g., `hours_NS`) are excluded from the calculation. Their values are **subtracted** from the total population to avoid skewing the average.
   - This ensures that only people with valid hours worked are considered when calculating the average hours worked.


4. **Final Average Calculation**:
   - The **weighted sum of hours worked** is divided by the **adjusted total population** (i.e., after subtracting individuals with 0 hours worked or not stated). This yields the weighted average hours worked for each profession.



```python
#Calculates the average hours worked of each occupation for Male, Female and Persons.
# Stores this average hour data as a dictionary
def calculate_weighted_avg_by_occupation(professions_dict):
    occupation_averages = {}

    for occupation, df in professions_dict.items():
        total_male = 0
        total_female = 0
        total_persons = 0
        tot_row = df[df['Hours_worked'] == 'Tot']
        if tot_row.empty:
            continue
        
        tot_male = tot_row['Male'].values[0]
        tot_female = tot_row['Female'].values[0]
        tot_persons = tot_row['Person'].values[0]
        tot_male_extra = 0
        tot_female_extra = 0
        tot_persons_extra = 0
        
        # Loops through each row except the "Tot" row
        for _, row in df[df['Hours_worked'] != 'Tot'].iterrows():
            hours_worked = row['Hours_worked']
            median_hours = get_median(hours_worked)
            
            if median_hours is None or median_hours == 0:
                tot_male_extra += row['Male']
                tot_female_extra += row['Female']
                tot_persons_extra += row['Person']
                continue
            
            if median_hours>50:
                median_hours = 60
                
            # Multiplies Male, Female, and Person by the median of the hours worked
            total_male += row['Male'] * median_hours
            total_female += row['Female'] * median_hours
            total_persons += row['Person'] * median_hours

        # Subtracts the hours_NS and the 0 hours worked row values from the total population 
        tot_male -= tot_male_extra
        tot_female -= tot_female_extra
        tot_persons -= tot_persons_extra
        
        # Calculates averages by dividing by the total "Tot" value
        avg_male = total_male / tot_male if tot_male != 0 else 0
        avg_female = total_female / tot_female if tot_female != 0 else 0
        avg_persons = total_persons / tot_persons if tot_persons != 0 else 0
        
        # Store the results in the dictionary
        occupation_averages[occupation] = [avg_male, avg_female, avg_persons]

    return occupation_averages

#HELPER FUNCTION
#Cleans and computes the median of the 'Hours worked' column by handling ranges and other string patterns.
def get_median(hours_worked):
    # Handle hours ranges like "1_19" or "1-9"
    if "NS" in hours_worked:  # Handle cases like "hours_NS"
        return None  # This will ignore the case of not stated hours
    elif "50_over" in hours_worked or "hours" in hours_worked:  # Handle cases like "50_over" or "60 hours or more worked" are taken as
        return 54.5  
    elif "_" in hours_worked:
        try:
            low, high = map(int, hours_worked.split("_"))
            return (low + high) / 2
        except ValueError:
            return None  
    elif "-" in hours_worked:  
        try:
            low, high = map(int, hours_worked.split("-"))
            return (low + high) / 2
        except ValueError:
            return None  
    elif hours_worked.isdigit():  
        return int(hours_worked)
    return None  
```

### Creating a Gender Percentage Dataframe

```python
# Creates a gender_percentage datafram that contains value for each gender
def calculate_gender_percentage(profession_dfs):
    data = []
    for profession, df in profession_dfs.items():
        total_row = df[df['Hours_worked'] == 'Tot']
        if not total_row.empty:
            total_male = total_row['Male'].values[0]
            total_female = total_row['Female'].values[0]
            total_person = total_row['Person'].values[0]
            male_percentage = (total_male / (total_male + total_female)) * 100 if total_person != 0 else 0
            female_percentage = (total_female / (total_male + total_female)) * 100 if total_person != 0 else 0
            data.append({
                'Occupation': profession,
                'Male_Percentage': male_percentage,
                'Female_Percentage': female_percentage
            })
    percentage_df = pd.DataFrame(data)
    return percentage_df
```

```python
# Calculates the employment rates
def calculate_tot_values(nz_dict, aus_dict):
    TotAus_Male, TotAus_Female, TotNZ_Male, TotNZ_Female = 0, 0, 0, 0
    nz_tot_row = nz_dict[nz_dict['Hours_worked'] == 'Total people stated'].iloc[0]
    aus_tot_row = aus_dict[aus_dict['Hours_worked'] == 'Tot'].iloc[0]
    #Calculating the values
    TotAus_Male = (aus_tot_row['Male'] * 100) / aus_tot_row['Person']
    TotAus_Female = (aus_tot_row['Female'] * 100) / aus_tot_row['Person']
    TotNZ_Male = (nz_tot_row['Male'] * 100) / nz_tot_row['Person']
    TotNZ_Female = (nz_tot_row['Female'] * 100) / nz_tot_row['Person']
    return TotAus_Male, TotAus_Female, TotNZ_Male, TotNZ_Female

# Plots the employment rates
def plot_employment_rates(TotAus_Male, TotAus_Female, TotNZ_Male, TotNZ_Female):
    values_male = [TotAus_Male, TotNZ_Male]  
    values_female = [TotAus_Female, TotNZ_Female] 
    categories = ['Australia', 'New Zealand']
    x = np.arange(len(categories))
    width = 0.2
    fig, ax = plt.subplots()
    bars_male = ax.bar(x - width/2, values_male, width, label='Males', color='blue')
    bars_female = ax.bar(x + width/2, values_female, width, label='Females', color='pink')
    for bar in bars_male:
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.1f}%', ha='center', va='bottom')
    for bar in bars_female:
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.1f}%', ha='center', va='bottom')

    # Setting labels and title
    ax.set_ylabel('Percentage (%)')
    ax.set_title('Percentage of Males and Females in Aus and NZ')
    ax.set_xticks(x)
    ax.set_xticklabels(categories)
    ax.set_ylim(0, 100)
    ax.legend()
    plt.tight_layout()
    plt.show()

TotAus_Male, TotAus_Female, TotNZ_Male, TotNZ_Female = calculate_tot_values(nz_total, aus_total)
plot_employment_rates(TotAus_Male, TotAus_Female, TotNZ_Male, TotNZ_Female)
```

### KEY FINDINGS
- In both countries, females constitute a **smaller percentage** of the workforce compared to males.
- **Australia** has a slightly higher percentage of women in the workforce compared to **New Zealand** (by about **1.1%**).
- This gender imbalance is consistent across both countries

### Total number of males and females working in different occupations

```python
# Plots the total number of males and females working in each different occupations
def plot_total_male_female_by_tot(profession_dfs, country_name, profession_order):
    professions = []
    total_males = []
    total_females = []

    for profession, df in profession_dfs.items():
        tot_row = df[df['Hours_worked'] == 'Tot']
        if not tot_row.empty:
            total_male = tot_row['Male'].values[0]
            total_female = tot_row['Female'].values[0]
            professions.append(profession)
            total_males.append(total_male)
            total_females.append(total_female)

    ordered_indices = [professions.index(prof) for prof in profession_order]
    ordered_males = [total_males[i] for i in ordered_indices]
    ordered_females = [total_females[i] for i in ordered_indices]
    
    bar_width = 0.35
    index = np.arange(len(profession_order))  
    plt.figure(figsize=(10, 6))
    plt.bar(index, ordered_males, bar_width, label='Male', color='blue')
    plt.bar(index + bar_width, ordered_females, bar_width, label='Female', color='pink')
    
    # Setting labels and title
    plt.xlabel('Occupations')
    plt.ylabel('Total People (in hundreds of thousands)')
    plt.title(f'{country_name}: Total Males and Females working in Each Occupation')
    
    # Setting y-axis with denominations of 100,000
    max_value = max(ordered_males + ordered_females)
    max_y = (int(np.ceil(max_value / 100000)) + 1) * 100000  
    y_ticks = np.arange(0, max_y + 100000, 100000)  
    plt.ticklabel_format(style='plain', axis='y')  
    
    plt.yticks(y_ticks, [f'{tick:,}' for tick in y_ticks])  
    plt.xticks(index + bar_width / 2, profession_order, rotation=45, ha='right')
    plt.legend()
    plt.figure(figsize=(12, 8))
    plt.tight_layout()
    plt.show()

# Creates a profession order 
profession_order = sorted(aus_professions_dict.keys())
```

```python
# Plots 2 bar graphs depicting total males and females working within each occupation for both countries
plot_total_male_female_by_tot(aus_professions_dict, 'Australia', profession_order)
plot_total_male_female_by_tot(nz_professions_dict, 'New Zealand', profession_order)
```

- Conversely, women are predominantly employed as **Professionals, Community and Personal Service Workers, Sales Workers, and Clerical and Administrative Workers**.

### Average hours worked by each gender based on occupation

```python

#Plots side by side radar charts
def plot_radar_chart_side_by_side(aus_averages, nz_averages):
    

    labels = list(occupation_short_forms.keys())  
    labels_short = [occupation_short_forms[occupation] for occupation in labels]
    aus_avg_male = [aus_averages[occupation][0] for occupation in labels]
    aus_avg_female = [aus_averages[occupation][1] for occupation in labels]
    nz_avg_male = [nz_averages[occupation][0] for occupation in labels]
    nz_avg_female = [nz_averages[occupation][1] for occupation in labels]
    num_vars = len(labels_short)
    # Compute angle for each occupation on the radar chart (in radians)
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

    # Make the plot circular by appending the first value to the end
    aus_avg_male += aus_avg_male[:1]
    aus_avg_female += aus_avg_female[:1]
    nz_avg_male += nz_avg_male[:1]
    nz_avg_female += nz_avg_female[:1]
    angles += angles[:1]

    # Initialize the radar charts with two subplots (side by side)
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8), subplot_kw=dict(polar=True))

    # Plot for Australia (ax1)
    ax1.fill(angles, aus_avg_male, color='blue', alpha=0.25, label='Avg Male (Australia)')
    ax1.plot(angles, aus_avg_male, color='blue', linewidth=2)
    ax1.fill(angles, aus_avg_female, color='pink', alpha=0.25, label='Avg Female (Australia)')
    ax1.plot(angles, aus_avg_female, color='purple', linewidth=2)
    ax1.set_xticks(angles[:-1])
    ax1.set_xticklabels(labels_short, fontsize=12)
    ax1.set_ylim(0, max(max(aus_avg_male), max(aus_avg_female)) + 15)
    ax1.set_title("Australia: Male and Female Averages by Occupation", size=16, color='black', y=1.1)
    ax1.legend(loc='upper right')
    for i, angle in enumerate(angles[:-1]):
        ax1.text(angle, aus_avg_male[i] + 1, f'{aus_avg_male[i]:.1f}', color='blue', size=10, horizontalalignment='center')
        ax1.text(angle, aus_avg_female[i] + 1, f'{aus_avg_female[i]:.1f}', color='purple', size=10, horizontalalignment='center')

    # Plot for New Zealand (ax2)
    ax2.fill(angles, nz_avg_male, color='blue', alpha=0.25, label='Avg Male (New Zealand)')
    ax2.plot(angles, nz_avg_male, color='blue', linewidth=2)
    ax2.fill(angles, nz_avg_female, color='pink', alpha=0.25, label='Avg Female (New Zealand)')
    ax2.plot(angles, nz_avg_female, color='purple', linewidth=2)
    ax2.set_xticks(angles[:-1])
    ax2.set_xticklabels(labels_short, fontsize=12)
    ax2.set_ylim(0, max(max(nz_avg_male), max(nz_avg_female)) + 12)
    ax2.set_title("New Zealand: Male and Female Averages by Occupation", size=16, color='black', y=1.1)
    ax2.legend(loc='upper right')
    for i, angle in enumerate(angles[:-1]):
        ax2.text(angle, nz_avg_male[i] + 1, f'{nz_avg_male[i]:.1f}', color='blue', size=10, horizontalalignment='center')
        ax2.text(angle, nz_avg_female[i] + 1, f'{nz_avg_female[i]:.1f}', color='purple', size=10, horizontalalignment='center')

    plt.figtext(0.5, -0.12, "Mng: Managers,    Pro: Professionals,    TTW: Technicians and Trades Workers,    CA: Clerical and Administrative Workers,", ha='center', fontsize=12)
    plt.figtext(0.5, -0.16, "    CPS: Community and Personal Service Workers,    Sal: Sales Workers,    MOD: Machinery Operators and Drivers,    Lab: Labourers", ha='center', fontsize=12)

    # Displays the radar charts
    plt.tight_layout()
    plt.show()

plot_radar_chart_side_by_side(aus_averages, nz_averages)
```

### KEY FINDINGS
- In every occupation, **women work fewer hours than men**, even in female-dominated roles.
- The difference in hours worked is particularly **significant for Technicians and Trades Workers** and **Sales Workers**.
- The patterns of average hours worked are similar in both countries.

## Results

From the above analysis, it can be concluded that:

1. Occupational Segregation is Widespread:

    * Both Australia and New Zealand exhibit clear patterns of occupational segregation where men and women are concentrated in different job types.

    * Men dominate traditionally male-dominated roles such as Technicians and Trades Workers, Machinery Operators, and Labourers. In contrast, women are predominantly found in roles such as Professionals, Clerical and Administrative Workers, and Community and Personal Service Workers.
    
    * This segregation suggests that gender plays a significant role in determining the types of jobs people enter, reinforcing traditional gender norms and potentially limiting opportunities for both genders in certain fields.

2. Gender Disparity in Work Hours:

    * Across almost all occupations, women work fewer hours than men, even in professions where they make up the majority of the workforce. This trend is especially visible in Sales Workers, Professionals, and Community and Personal Service Workers.
      
    * The disparity in hours worked contributes to wage differences, as working fewer hours generally correlates with lower total income, reinforcing gender-based economic inequalities.

      
3. Female-Dominated Occupations Are Not Immune to Inequality:
   
    * While women are overrepresented in roles such as Clerical Workers, Sales, and Personal Service, they still tend to work fewer hours compared to men. This points to systemic factors, such as societal expectations or caregiving responsibilities, limiting women's availability for full-time or extended-hour positions.
      
    * Furthermore, many female-dominated roles, such as Sales Workers and Community Services, often tend to be lower-paying professions, contributing to the gender wage gap.
      
5. Consistent Trends Across Both Countries:
   
    * The patterns observed are remarkably consistent between Australia and New Zealand. While Australia has a slightly higher percentage of female workforce participation, the overall distribution of genders across occupations and the differences in hours worked remain very similar in both countries. This consistency suggests that broader cultural or societal factors, rather than country-specific policies alone, contribute to these gender disparities in the labor force.
